```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

countries = ["China", "India", "USA", "Indonesia", "Pakistan"]
population = [1420000000, 1400000000, 339000000, 277000000, 240000000]

plt.figure(figsize=(10,6))
bars = plt.bar(countries, population, color="skyblue", edgecolor="black")

formatter = ticker.FuncFormatter(lambda x, _: f'{x/1e9:.1f}B' if x >= 1e9 else f'{x/1e6:.1f}M')
plt.gca().yaxis.set_major_formatter(formatter)

plt.title("Top 5 Most Populous Countries", fontsize=16, fontweight="bold")
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)

for bar in bars:
    h = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, h,
             f'{h/1e9:.2f}B' if h >= 1e9 else f'{h/1e6:.2f}M',
             ha='center', va='bottom', fontsize=10, fontweight="bold")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

genres = ["Action", "Drama", "Comedy", "Romance", "Horror", "Thriller", "Sci-Fi"]
movie_count = [950, 1200, 1100, 700, 500, 800, 600]

plt.figure(figsize=(10,6))
bars = plt.bar(genres, movie_count, color="lightcoral", edgecolor="black")

plt.title("Number of Movies Produced by Genre", fontsize=16, fontweight="bold")
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.xticks(rotation=45)

for bar in bars:
    h = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, h, str(h), ha='center', va='bottom', fontsize=10, fontweight="bold")

plt.tight_layout()
plt.show()
```
